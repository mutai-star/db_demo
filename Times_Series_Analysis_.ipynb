```python
from pycaret.datasets import get_data

# Load the dataset
data = get_data('airline')
print(data.head())
```

#Data Preprocessing
#fh=12 specifies the forecast horizon (12 months in this case)



```python
from pycaret.time_series import *

# Initialize the time series experiment
exp = setup(data, fh=12, session_id=123)
```

Data Visualization
time series
decomposition and autocorellation

```python
# Plot the time series
plot_model(plot='ts')

# Decompose the time series into trend, seasonality, and residuals
plot_model(plot='decomp')

# Plot autocorrelation
plot_model(plot='acf')
```

# Implementation and Evaluation of time series models;
1. ARIMA
2. Exponential smoothing
3. Prophet

```python
# Create and compare models
best_model = compare_models (sort='MAE')

# Train specific models
arima = create_model('arima')
ets = create_model('ets')
prophet = create_model('prophet')

# Evaluate models using visual diagnostics
plot_model(arima, plot='forecast')  # Forecast plot
plot_model(ets, plot='forecast')    # Forecast plot
plot_model(prophet, plot='forecast')  # Forecast plot

# Additional diagnostics (e.g., residuals, ACF)
plot_model(arima, plot='residuals')  # Residuals plot
plot_model(ets, plot='residuals')    # Residuals plot
plot_model(prophet, plot='residuals')  # Residuals plot
```

#Generating prediction with the best performance model

```python
# Finalize the best model
final_model = finalize_model(best_model)

# Generate predictions
predictions = predict_model(final_model)
print(predictions)

# Save the trained model
# Replace tuned_model with final_model
save_model(final_model, 'multi_step_forecast_model')
```
