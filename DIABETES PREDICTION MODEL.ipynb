```python
# Enhanced Data Mining Models for Google Colab
# Classification, Clustering, and Association Rule Mining

# Install required packages
!pip install mlxtend --quiet

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from mlxtend.frequent_patterns import apriori, association_rules
from itertools import combinations
from collections import defaultdict
import requests
import io
from google.colab import files
import warnings
warnings.filterwarnings('ignore')

# Set matplotlib style
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 6)

# Dataset URLs for Colab
DATASETS = {
    'diabetes': 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv',
    'mall_customers': 'https://raw.githubusercontent.com/sahil3011/Mall-Customer-Segmentation/master/Mall_Customers.csv',
    'groceries': 'https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/groceries.csv'
}

def load_dataset(url):
    """Load dataset directly from URL into DataFrame"""
    response = requests.get(url)
    response.raise_for_status()
    return pd.read_csv(io.StringIO(response.text))

def explore_data(df, name):
    """Display basic dataset info and visualizations"""
    print(f"\n=== {name.upper()} DATASET ===")
    display(df.head())

    # Basic statistics
    print("\nüìä Basic Statistics:")
    display(df.describe())

    # Visualizations
    if name == 'diabetes':
        plt.figure(figsize=(10, 6))
        df.hist()
        plt.tight_layout()
        plt.show()
    elif name == 'mall_customers':
        sns.pairplot(df.drop('CustomerID', axis=1))
        plt.show()

# 1. Diabetes Classification Model (Enhanced)
def diabetes_analysis():
    print("\nüîÆ DIABETES PREDICTION MODEL")
    try:
        # Load and prepare data
        df = load_dataset(DATASETS['diabetes'])
        df.columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
                     'Insulin', 'BMI', 'DiabetesPedigree', 'Age', 'Outcome']

        explore_data(df, 'diabetes')

        # Handle missing values (zeros)
        df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = \
            df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0, np.nan)
        df.fillna(df.mean(), inplace=True)

        # Feature engineering
        df['Glucose_BP'] = df['Glucose'] * df['BloodPressure']
        df['BMI_Age'] = df['BMI'] * df['Age']

        X = df.drop('Outcome', axis=1)
        y = df['Outcome']

        # Train-test split with stratification
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.25, random_state=42, stratify=y)

        # Model training
        model = LogisticRegression(max_iter=1000, class_weight='balanced')
        model.fit(X_train, y_train)

        # Evaluation
        y_pred = model.predict(X_test)
        print("\nüéØ Model Performance:")
        print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
        print("\nüìù Classification Report:")
        print(classification_report(y_test, y_pred))

        # Confusion matrix visualization
        plt.figure(figsize=(8, 6))
        sns.heatmap(confusion_matrix(y_test, y_pred),
                    annot=True, fmt='d', cmap='Blues',
                    xticklabels=['No Diabetes', 'Diabetes'],
                    yticklabels=['No Diabetes', 'Diabetes'])
        plt.title('Confusion Matrix')
        plt.show()

        # Feature importance
        importance = pd.DataFrame({
            'Feature': X.columns,
            'Importance': model.coef_[0]
        }).sort_values('Importance', ascending=False)

        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=importance)
        plt.title('Feature Importance')
        plt.show()

    except Exception as e:
        print(f"‚ùå Error in diabetes analysis: {str(e)}")

# 2. Customer Segmentation (Enhanced)
def customer_segmentation():
    print("\nüõçÔ∏è CUSTOMER SEGMENTATION MODEL")
    try:
        df = load_dataset(DATASETS['mall_customers'])
        explore_data(df, 'mall_customers')

        # Prepare data
        X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
        X_scaled = StandardScaler().fit_transform(X)

        # Determine optimal clusters
        wcss = []
        for i in range(1, 11):
            kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
            kmeans.fit(X_scaled)
            wcss.append(kmeans.inertia_)

        # Elbow plot
        plt.figure(figsize=(10, 6))
        plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
        plt.title('Elbow Method for Optimal k')
        plt.xlabel('Number of Clusters')
        plt.ylabel('WCSS')
        plt.show()

        # Final clustering
        kmeans = KMeans(n_clusters=5, random_state=42)
        df['Cluster'] = kmeans.fit_predict(X_scaled)

        # Visualization
        plt.figure(figsize=(12, 8))
        sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)',
                        hue='Cluster', palette='viridis', s=100)
        plt.title('Customer Segments', fontsize=16)
        plt.xlabel('Annual Income (k$)', fontsize=14)
        plt.ylabel('Spending Score (1-100)', fontsize=14)

        # Plot centroids
        centers = kmeans.cluster_centers_
        plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
        plt.show()

        # Cluster profiles
        print("\nüìä Cluster Profiles:")
        display(df.groupby('Cluster').agg({
            'Annual Income (k$)': ['mean', 'median'],
            'Spending Score (1-100)': ['mean', 'median'],
            'CustomerID': 'count'
        }).rename(columns={'CustomerID': 'Count'}))

    except Exception as e:
        print(f"‚ùå Error in customer segmentation: {str(e)}")

# 3. Market Basket Analysis (Enhanced)
def market_basket_analysis():
    print("\nüõí MARKET BASKET ANALYSIS")
    try:
        # Load data
        response = requests.get(DATASETS['groceries'])
        transactions = [line.strip().split(',') for line in response.text.split('\n') if line]

        # Item frequency analysis
        all_items = [item for transaction in transactions for item in transaction]
        item_counts = pd.Series(all_items).value_counts().reset_index()
        item_counts.columns = ['Item', 'Count']

        print("\nüìä Top 20 Frequently Purchased Items:")
        plt.figure(figsize=(12, 8))
        sns.barplot(x='Count', y='Item', data=item_counts.head(20))
        plt.title('Top 20 Most Frequent Items')
        plt.show()

        # Association rules
        from mlxtend.preprocessing import TransactionEncoder
        te = TransactionEncoder()
        te_ary = te.fit(transactions).transform(transactions)
        df = pd.DataFrame(te_ary, columns=te.columns_)

        # Generate frequent itemsets
        frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True, max_len=2)

        # Generate rules
        rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

        # Filter and sort rules
        strong_rules = rules[(rules['lift'] >= 1.5) & (rules['confidence'] >= 0.3)]\
            .sort_values('lift', ascending=False)

        print("\nüîó Strong Association Rules (lift ‚â• 1.5, confidence ‚â• 0.3):")
        display(strong_rules.head(10))

        # Visualization
        plt.figure(figsize=(12, 8))
        sns.scatterplot(data=strong_rules, x='support', y='confidence', size='lift', hue='lift', sizes=(50, 200))
        plt.title('Association Rules (Size = Lift)', fontsize=16)
        plt.show()

    except Exception as e:
        print(f"‚ùå Error in market basket analysis: {str(e)}")

# Main execution
if __name__ == "__main__":
    print("üöÄ Starting Data Mining Models in Google Colab")

    # Run all analyses
    diabetes_analysis()
    customer_segmentation()
    market_basket_analysis()

    print("\n‚úÖ All analyses completed!")
```
